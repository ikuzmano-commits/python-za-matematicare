```
def f_string(lista):
    dulj=0
    for x in lista:
        if len(x)>dulj:
            dulj=len(x)
            najdulji=x
    print('Duljina je:', dulj)
    print('Tra≈æeni string je:', najdulji)
    poz = open('pozivi.txt','a')
    poz.write('Argument je: '+str(lista)+'\n')
    poz.write('Rezultat je: '+str(dulj)+' ' +str(najdulji)+'\n')
    poz.close()
    return (dulj,najdulji)
```

```
lista=['ponedjeljak','utorak','srijeda','cetvrtak','petak','subota','nedjelja']
f_string(lista)
lista1=['aa','bbb','ccccccccc','dd','eeeeee']
f_string(lista1)
```
