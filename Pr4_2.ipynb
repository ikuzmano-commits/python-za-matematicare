```
lista = ['prvi', 'a', 4, -4.6, True]
```

```
print(len(lista))      # duljina liste (broj elemenata u listi)
```

```
print(lista[1:3])      # lista[a:b] sadrži elemente s indeksima od a do b-1
print(lista[1:-1])     # elementi od indeksa 1 do pretposljednjeg
print(lista[:3])       # od početka do elementa s indeksom 2
print(lista[3:])       # od elementa s indeksom 3 do kraja
```
