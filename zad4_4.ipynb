```
lista = [5, 6, 8, 1, 9, 11, 12, 3]
```

```
# Najmanji element
minel = lista[0]
for k in lista:
    if k < minel:
        minel = k
print("Najmanji element u listi je:", minel)
```

```
# Drugi najveći element
# prvo pronalazimo najveći element
maxel = lista[0]
for k in lista:
    if k > maxel:
        maxel = k
```

```
# uklanjamo najveći element iz liste
lista.remove(maxel)
```

```
# ponovno tražimo najveći —  to je drugi po veličini u početnoj listi
drugimax = lista[0]
for k in lista:
    if k > drugimax:
        drugimax = k
print("Drugi najveći element u listi je:", drugimax)
```
