```
import random
a=random.randint(1,20) # slučajan broj između 1 i 20  
p=0 # postavljanje početne vrijednosti
```

```
while p != a:
    p=int(input("Unesite broj između 1 i 20: "))

    if p > a:
        print("Broj je manji od unešenog.")
    elif p < a:
        print("Broj je veći od unešenog.")
    else:
        print("Pogodak!")
```
