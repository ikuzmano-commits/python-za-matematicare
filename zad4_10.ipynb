```
def listasort(lista, smjer):
    n = len(lista)
    sortirana = []
    lista1=lista.copy()   #nećemo promjene raditi na originalnoj listi
    
    if smjer == 'uzlazno':
        for i in range(n):
            najmanji = lista1[0]
            for x in lista1:
                if x < najmanji:
                    najmanji = x
            lista1.remove(najmanji)
            sortirana.append(najmanji)
    
    elif smjer == 'silazno':
        for i in range(n):
            najveci = lista1[0]
            for x in lista1:
                if x > najveci:
                    najveci = x
            lista1.remove(najveci)
            sortirana.append(najveci)
    else:
        print("Greška: drugi argument mora biti 'uzlazno' ili 'silazno'.")
        return None
    
    return sortirana  
```

```
brojevi = [5, 6, 8, 9, 11, 12, 3]
print("Uzlazno sortirano:", listasort(brojevi, 'uzlazno'))
print("Silazno sortirano:", listasort(brojevi, 'silazno'))
```
