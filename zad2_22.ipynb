```
import random
broj=0
b1=0
b2=0
b3=0
b4=0
b5=0
```

```
while broj!=6:
    broj=random.randrange(1,7)
    if broj==1:
        b1=b1+1
    if broj==2:
        b2=b2+1
    if broj==3:
        b3=b3+1
    if broj==4:
        b4=b4+1
    if broj==5:
        b5=b5+1
    print(broj)
```

```
print(f'Broj 1 je pao {b1} puta')
print(f'Broj 2 je pao {b2} puta')
print(f'Broj 3 je pao {b3} puta')
print(f'Broj 4 je pao {b4} puta')
print(f'Broj 5 je pao {b5} puta')
```
