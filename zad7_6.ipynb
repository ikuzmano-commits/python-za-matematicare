```
import numpy as np
```

```
def read_matrices(filename):
    matrice = []
    f = open(filename, "r")
    current = []

    for line in f:
        line = line.strip()

        if line == "":
            if current != []:
                matrice.append(np.array(current))
                current = []
        else:
            row = []
            for x in line.split():
                row.append(float(x))
            current.append(row)

    if current != []:
        matrice.append(np.array(current))

    f.close()
    return matrice
```

```
def simetricna(matrix):
    return np.array_equal(matrix, matrix.T)
```

```
def antisimetricna(matrix):
    return np.array_equal(matrix, -matrix.T)
```

```
# glavni dio
matrice = read_matrices("matrice.txt")
sim_matr = []
```

```
for  mat in matrice:
    if simetricna(mat):
        sim_matr.append(mat)
```

```
# zapis simetričnih matrica u datoteku
f=open("simetricne.txt", "w")
for mat in sim_matr:
   for row in mat:
      f.write(" ".join(str(int(x)) if x.is_integer() else str(x) for x in row) + "\n")
   f.write("\n")  # prazna linija između matrica
f.close()
```
