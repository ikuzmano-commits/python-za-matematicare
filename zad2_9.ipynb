```
# unos broja s provjerom
n = int(input('Unesite broj: '))
```

```
if n <= 0:
    print("Molim vas, unesite pozitivan cijeli broj.")
else:
    print("Parni brojevi od 1 do", n, ":")
     # ispis parnih brojeva
    for i in range(2, n+1, 2):
        print(i)
```

```
# alternativna metoda 1
for i in range(1, n+1):
    if i % 2 == 0:
        print(i)
```

```
# alternativna metoda 2
for i in range(1,int(n/2)+1):
    print(2*i)
```
