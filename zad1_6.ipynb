```python
# koeficijenti kvadratne jednadÅ¾be
a = float(input('Unesi a: '))
b = float(input('Unesi b: '))
c = float(input('Unesi c: '))
```

```python
# izračun diskriminante
d = b**2 - 4*a*c  # diskriminanta
```

```python
#rjeÅ¡enja
x1 = (-b + d**0.5) / (2 * a)
x2 = (-b - d**0.5) / (2 * a)
```

```python
# ispis rezultata - {x1:.2f} i {x2:-2f} omogućava ispis s dva decimalna mjesta
print(f'Rješenja kvadratne jednadžbe {a}x^2 + {b}x + {c} su x1 = {x1:.2f} i x2 = {x2:.2f}')
```
