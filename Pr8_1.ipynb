```
#Definiranje klase
class Pravokutnik:
    "Pravokutnik s metodama za povrsinu, opseg i skaliranje"
    # konstruktor koji se poziva automatski kad se stvori objekt klase
    def __init__(self,x,y):
    # x,y predstavljaju duljine stranica pravokutnika
        self.x=x
        self.y=y
    # metoda za računanje površine
    def povrsina(self):
        return self.x*self.y
    # metoda za računanje opsega
    def opseg(self):
        return 2*(self.x+self.y)
    # metoda za skaliranje
    def skaliranje(self,k):
        self.x=self.x*k
        self.y=self.y*k
        print(self.x,self.y) # ispis novih x, y nakon skaliranja
```

```
# Korištenje klase
# stvaranje objekta P koji je pravokutnik s duljinama stranica 10 i 15
P=Pravokutnik(10,15)
```

```
print(P.povrsina()) # ispis površine pravokutnika P
print(P.opseg()) # ispis opsega praokutnika P
P.skaliranje(0.2) # skaliranje duljina stranica
print(P.povrsina()) # ispis površine pravokutnika P nakon skaliranja
```
