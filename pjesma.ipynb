```
Mlačna noć; u selu lavež; kasan
Ćuk il netopir;
Ljubav cvieća — miris jak i strasan
Slavi tajni pir.
```

```
Sitni cvrčak sjetno cvrči, jasan
Kao srebrn vir;
Težke oči sklapaju se na san,
S neba rosi mir.
```

```
S mrkog tornja bat
Broji pospan sat,
Blaga svjetlost sipi sa visina;
```
