```
import random
```

```
file_voce = open('voce.txt')
file_pokusaji = open('pokusaji.txt','w')
lista_voce = []
for line in file_voce:
    lista_voce.append(line.rstrip())
file_voce.close()
```

```
def anagram(lista):
    voce = random.choice(lista)
    pom_lista=list(voce)
    random.shuffle(pom_lista)
    anagram = ''.join(pom_lista)
    return (voce,anagram)
```

```
(voce, anagram) = anagram(lista_voce)
print('Anagram je: '+anagram)
file_pokusaji.write('Anagram je: '+anagram)
print('Od kog je voća nastao ovaj anagram? \n')
file_pokusaji.write('Od kog je voća nastao ovaj anagram? \n')
```

```
pogodena = ''
brojac = 1
while pogodena != voce:
    pogodena = input('voće: ')
    file_pokusaji.write(str(brojac)+'. '+pogodena+'\n')
    brojac+=1
print('Pogodili ste!')
file_pokusaji.write('Pogodili ste!')
```

```
file_pokusaji.close()
```
