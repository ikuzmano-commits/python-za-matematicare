```
import numpy as np
```

```
def podmatrice(A):
    n=A.shape[0]
    for i in range(1,n+1):
        B=A[:i,:i]
        trag=0
        for j in range(i):   #ili trag = np.trace(B)
            trag=trag+A[j,j]
        print(B)
        print('Trag je:',trag)
```

```
n=10
A=np.random.randint(-10,11, size=(n,n))
podmatrice(A)
```
