```
class Pravokutnik:
    "Pravokutnik s metodama za povrsinu, opseg i skaliranje"
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def povrsina(self):
        return self.x*self.y
    def opseg(self):
        return 2*(self.x+self.y)
    def skaliranje(self,k):
        self.x=self.x*k
        self.y=self.y*k
        print(self.x,self.y)
```

```
class Kvadrat(Pravokutnik):
    "Klasa kvadrat nasljedjuje klasu pravokutnik"
    def __init__(self, x):
        super(Kvadrat,self).__init__(x,x)
    def ispis(self):
        print(f"Zadali ste kvadrat sa stranicom duljine {self.x}") 
```

```
K=Kvadrat(2)
print(K.opseg())
K.ispis()
```
