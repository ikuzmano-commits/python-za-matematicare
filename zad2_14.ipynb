```
# unos broja članova Fibonaccijevog niza od korisnika
n = int(input("Unesite željeni n: "))
if n <0:
    print("Molimo unesite nenegativni cijeli broj.")
elif n == 0:
    print("Fibonaccijev niz: 0")
else:
    f0 = 0
    print(f0)
    f1 = 1
    print(f1)
    # petlja za izračun ostalih članova
    for i in range(1, n):
        f = f0 + f1
        print(f)
        # ažuriramo prethodne članove
        f0=f1
        f1=f
```
