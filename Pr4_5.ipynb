```
a = [2, 4, 5, 11]
```

```
# prolazak kroz elemente liste i ispis svakog uveÄ‡anog za 3
for x in a:
    print(x + 3)
```

```
# zbrajanje svih elemenata liste
suma = 0
for x in a:
    suma = suma + x
```

```
print("Suma elemenata liste je:", suma)
```

```
# sortiranje liste uzlazno
a.sort()
print("Sortirana lista (uzlazno):", a)
```

```
# sortiranje liste silazno
a.sort(reverse=True)
print("Sortirana lista (silazno):", a)
```
