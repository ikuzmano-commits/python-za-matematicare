```
lista = [1, 6, 8, 3, 1, 6, 6, 3]
```

```
# Ispis elemenata koji se pojavljuju više puta
print("Elementi koji se ponavljaju:")
n = len(lista)
for i in range(0, n):
    for j in range(i + 1, n):
        if lista[i] == lista[j]:
            print(lista[i])
            break   # stajemo traziti duplikat ako smo ga vec našli
```

```
# Uklanjanje duplikata i formiranje nove liste 
nova = []
for x in lista:
    if x not in nova:
        nova.append(x)
```

```
print("Lista bez duplikata:", nova)
```
