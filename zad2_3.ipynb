```
a = float(input("Unesite duljinu stranice a: "))
b = float(input("Unesite duljinu stranice b: "))
c = float(input("Unesite duljinu stranice c: "))
if a<=0 or b<=0 or c<=0:
    print('Nije dobar unos!')
else:
    if a == b == c:
        print("Trokut je jednakostraničan.")
    elif a == b or b == c or a == c:
        print("Trokut je jednakokračan.")
    else:
        print("Trokut je raznostraničan.")
```
