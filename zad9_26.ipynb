```
import turtle
```

```
t = turtle.Turtle()
t.hideturtle()
turtle.bgcolor("white")
```

```
# funkcija za trokut (grane)
def trokut(a):
    t.begin_fill()
    for i in range(3):
        t.forward(a)
        t.left(120)
    t.end_fill()
```

```
# crtanje boriÄ‡a
t.color("darkgreen")
```

```
# donji trokut
t.penup()
t.goto(-90, -100)
t.pendown()
trokut(180)
```

```
# srednji trokut
t.penup()
t.goto(-70, -40)
t.pendown()
trokut(140)
```

```
# gornji trokut
t.penup()
t.goto(-50, 20)
t.pendown()
trokut(100)
```

```
# deblo
t.color("saddlebrown")
t.penup()
t.goto(-15, -160)
t.pendown()
t.begin_fill()
for i in range(2):
    t.forward(30)
    t.left(90)
    t.forward(60)
    t.left(90)
t.end_fill()
```

```
# ----- KUGLICE -----
kuglice = [
    (-40, -40, "red"),
    (0, -20, "yellow"),
    (40, -40, "blue"),
    (-20, 10, "purple"),
    (20, 30, "orange")
]
```

```
for x, y, boja in kuglice:
    t.penup()
    t.goto(x, y)
    t.color(boja)
    t.dot(12)
```

```
turtle.done()
```
