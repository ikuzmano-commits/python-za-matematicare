```
import pandas as pd
import matplotlib.pyplot as plt
```

```
# učitavanje podataka
url = ("https://raw.githubusercontent.com/datasciencedojo/"
"datasets/master/titanic.csv"
)
data = pd.read_csv(url)
```

```
# osnovne informacije
print(data.shape)
print(data.head(5))
print(data.columns)
```

```
# filtriranje stupaca
procisceni_podaci = data[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare']]
print(procisceni_podaci.head(5))
```

```
# histogrami dobi i cijene karata
plt.subplot(2,1,1)
plt.hist(procisceni_podaci.Age.dropna())
plt.title('Distribucija dobi i cijene karata')
plt.xlabel('dob')
```

```
plt.subplot(2,1,2)
plt.hist(procisceni_podaci.Fare.dropna())
plt.xlabel('cijena karte')
plt.show()
```

```
# osnovna statistika i filtriranje
print("Putnika sa dob < 18:", len(procisceni_podaci[procisceni_podaci.Age < 18]))
print("Putnika sa cijena karte < 10:", len(procisceni_podaci[procisceni_podaci.Fare < 10]))
```

```
procisceni_podaci = procisceni_podaci[(procisceni_podaci.Age >= 18) & (procisceni_podaci.Fare >= 10)]
print(procisceni_podaci.shape)
```

```
# histogram cijene karata
plt.hist(procisceni_podaci.Fare.dropna())
plt.title('Cijena karata')
plt.show()
```

```
# mean i median dobi
prVrij = procisceni_podaci.Age.mean()
med = procisceni_podaci.Age.median()
print("Prosjek dobi:", prVrij, "Medijan dobi:", med)
```

```
# mlađi i stariji putnici
mladi_putnici = procisceni_podaci[procisceni_podaci.Age < prVrij]
stariji_putnici = procisceni_podaci[procisceni_podaci.Age >= prVrij]
```

```
# scatter plot
plt.subplot(2,1,1)
plt.scatter(stariji_putnici.Age, stariji_putnici.Fare, marker='o')
plt.title('Odnos dobi i cijene karte')
plt.ylabel('cijena karte')
```

```
plt.subplot(2,1,2)
plt.scatter(mladi_putnici.Age, mladi_putnici.Fare, marker='x')
plt.xlabel('dob')
plt.ylabel('cijena karte')
```
