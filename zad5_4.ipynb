```
import numpy as np
```

```
def najbliziAritmSred(v):
    arSr=np.mean(v)
    mind=abs(v[0][0]-arSr)
    element=v[0][0]
    for el in v[0]:
        if mind>abs(el-arSr):
            mind=abs(el-arSr)
            element=el
    print('Aritmeti훾ka sredina je ',arSr)
    print('element najbli탑i aritmeti훾koj sredini je',element)
```

```
n=30
v=np.random.randint(-25,45, size=(1,n)) #1D polje se mo탑e tretirati kao matrica s jednim retkom (ili stupcem)
print('v=',v)
najbliziAritmSred(v)
```
