```
a = int(input('Unesite prirodni broj a: '))
b = float(input('Unesite realan broj b: '))
# zamjena vrijednosti uvođenjem treće varijable
c=a
a=b   # isto se može napraviti kraće i kao a,b = b,a 
b=c
```

```
print(f'Nakon zamjene vrijednosti, varijable a i b imaju vrijednosti {a} i {b}.')
```

```
print(f'Nakon zamjene: a = {a} (tip: {type(a)}), b = {b} (tip: {type(b)})')
```
