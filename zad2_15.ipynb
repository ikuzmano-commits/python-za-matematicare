```
# unos znamenke jedinica od korisnika
jed = int(input("Unesite znamenku jedinica (0-9): "))
```

```
if jed < 0 or jed > 9:
    print("Molimo unesite znamenku u rasponu od 0 do 9.")
else:
    print(f"Dvoznamenkasti brojevi s znamenkom jedinica {jed} su:")

    # Petlja za prolazak kroz sve dvoznamenkaste brojeve
    for i in range(1, 10):
        broj = i * 10 + jed
        print(broj)
```
