```
import math
```

```
# unos broja podintervala
n = int(input("Unesite broj podintervala n: "))
```

```
# granice integracije
a = 0
b = math.pi
```

```
# širina podintervala
h = (b - a) / n
```

```
# početna suma sa sinusom na krajevima intervala
suma = (math.sin(a) + math.sin(b)) / 2
```

```
# dodavanje vrijednosti funkcije u unutarnjim točkama
for i in range(1, n):
    x_i = a + i * h
    suma += math.sin(x_i)
```

```
# trapezna aproksimacija
I = h * suma
```

```
# ispis rezultata
print(f"Aproksimirana vrijednost integrala je: {I:.8f}")
```

```
# točna vrijednost integrala
tocno = 2
greska = abs(I - tocno)
```

```
print(f"Apsolutna pogreška u odnosu na točnu vrijednost 2 iznosi: {greska:.8e}")
```
