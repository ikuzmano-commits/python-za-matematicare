```
voce = ['naranča', 'jabuka', 'kivi', 'kruška', 'ananas']
```

```
# dodavanje novog voća na kraj liste
voce.append('šljiva')
```

```
# uklanjanje jednog voća iz liste
voce.remove('kivi')
print(voce)
```

```
# provjera prisutnosti određenog voća u listi
if 'banana' in voce:
    print("Banana je u listi.")
else:
    print("Banana nije u listi.")
```

```
# sortiranje liste abecedno
voce.sort()
print(voce)
```

```
# ispis svake vrste voća i broja slova u nazivu
for v in voce:
    print(f"{v} ima {len(v)} slova")
```
