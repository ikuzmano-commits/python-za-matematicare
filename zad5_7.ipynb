```
import numpy as np
```

```
def dt2(A):
    return A[0][0]*A[1][1]-A[0][1]*A[1][0]
```

```
def minor(A, i, j):
    """
    Vraća matricu dobivenu izbacivanjem i-tog retka i j-tog stupca.
    """
    n = len(A)
    return [
        [A[r][c] for c in range(n) if c != j]
        for r in range(n) if r != i
    ]
```

```
def dtm(A):
    """
    Rekurzivno računa determinantu kvadratne matrice proizvoljnog reda
    razvojem po prvom retku.
    """
    n = len(A)

    if n == 1:
        return A[0][0]
    if n == 2:
        return dt2(A)

    det = 0
    # razvoj po prvom retku (i = 0)
    for j in range(n):
        det += ((-1) ** j) * A[0][j] * dtm(minor(A, 0, j))
    return det
```

```
A2 = np.array([[1, 2], [3, 4]])
print("dt2(A2) =", dt2(A2))  
print(np.linalg.det(A2)) #provjera
```

```
A3 = np.array([[2, 1, 3], [0, -3, 4], [15, 2, 0]])
print("dtm(A3) =", dtm(A3))    
print(np.linalg.det(A3)) #provjera
```
