```
def skracivanje(a, b):
    # provjera da nazivnik nije nula
    if b == 0:
        return("Nazivnik ne može biti nula.")

    # koristimo apsolutne vrijednosti jer NZD mora biti pozitivan
    for i in range(min(abs(a), abs(b)), 1, -1):
        if a % i == 0 and b % i == 0:
            a = a // i
            b = b // i
            break  # nakon što smo skratili, možemo izaći iz petlje

    # osiguravamo da je nazivnik pozitivan
    if b < 0:
        a = -a
        b = -b

    return a,b
```

```
# primjer poziva
(b1,n1) = skracivanje(34, -12)
print(f"Skraćeni razlomak je: {b1}/{n1}")
```
