```
# unos cijelog broja
broj = int(input("Unesite neki cijeli broj: "))
```

```
# provjera je li broj manji ili jednak 0
if broj <= 0:
    if broj == 0:
        print("Unesena vrijednost je nula.")
    else:
        kvadrat = broj ** 2
        print(f"Uneseni broj je negativan. Kvadrat tog broja je: {kvadrat}.")
else:
    # ako je broj između 1 i 10 (uključujući oba)
    if 1 <= broj <= 10:
        kvadrat = broj ** 2
        print(f"Broj je pozitivan i u rasponu između 1 i 10. Kvadrat broja je: {kvadrat}.")
    # Ako je broj između 11 i 100 (uključujući oba)
    elif 11 <= broj <= 100:
        razlika = broj - 25
        print(f"Broj je u rasponu od 11 do 100. Razlika broja i 25 je: {razlika}.")
        if razlika < 0:
            kvadrat_razlike = razlika ** 2
            print(f"Razlika je negativna. Kvadrat razlike je: {kvadrat_razlike}.")
        else:
            print(f"Razlika je pozitivna: {razlika}.")
    # Ako je broj veći od 100
    else:
        print("Uneseni broj je veći od 100.")
```
