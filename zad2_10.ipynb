```
n = int(input("Unesite broj n: "))
```

```
#Koristimo petlju for 
suma = 0 # inicijalnizacija sume
for i in range(1, n + 1):
    suma += i
print(f"Suma prvih {n} prirodnih brojeva je: {suma}")
```

```
#Koristiomo petlju while 
suma = 0 # inicijalizacija sume i broja
i = 1
while i <= n:
    suma += i
    i += 1
print(f"Suma prvih {n} prirodnih brojeva je: {suma}")
```
