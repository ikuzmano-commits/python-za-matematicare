```
import turtle
turtle.title('Crtanje spirale')
kornjaca = turtle.Turtle()
kornjaca.speed(10)
duljina=2
for _ in range(100):
    kornjaca.forward(duljina)
    kornjaca.left(90)  # Rotacija za 90 stupnjeva (pravokutna spirala)
    duljina += 2  # Uvećanje duljine stranice

    # Završi crtanje
kornjaca.hideturtle()
print("Crtanje završeno.")  # Informacija da je proces završen
turtle.done()
```
