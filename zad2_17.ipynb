```
br = 0 #inicijalna vrijednost brojača
```

```
# petlja koja prolazi kroz sve četveroznamenkaste brojeve sa zadnjom znamenkom 7
for broj in range(1007, 10000, 10):  # Počinjemo od 1007 i povećavamo za 10 kako bi zadnja znamenka uvijek bila 7
    if broj % 3 == 0:
        br += 1
```

```
# ispis broja pronađenih brojeva
print(f"Četveroznamenkastih brojeva koji su djeljivi s 3 i imaju zadnju znamenku 7 ima {br}.")
```
