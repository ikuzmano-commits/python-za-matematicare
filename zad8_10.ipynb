```
class Osoba:
    def __init__(self, ime, prezime, spol, godine):
        self.ime = ime
        self.prezime = prezime
        self.spol = spol
        self.godine = godine

    def pozdrav(self):
        print(f"Dobar dan! Ja sam {self.ime} {self.prezime} i imam {self.godine} godina.")

    def __ge__(self, o2):
        return self.godine > o2.godine
```

```
class Student(Osoba):
    def __init__(self, ime, prezime, spol, godine, godst, fakultet):
        super().__init__(ime, prezime, spol, godine)
        self.godst = godst
        self.fakultet = fakultet

    def polozioBiKolegij(self, kolegij, bodovi):
        if self.spol == 'Z' and bodovi >= 45:
            print(f"Studentica {self.ime} {self.prezime} položila je kolegij {kolegij} na fakultetu {self.fakultet}.")
        elif self.spol == 'Z' and bodovi < 45:
            print(f"Studentica {self.ime} {self.prezime} nije položila kolegij {kolegij} na fakultetu {self.fakultet}.")
        elif self.spol == 'M' and bodovi >= 45:
            print(f"Student {self.ime} {self.prezime} položio je kolegij {kolegij} na fakultetu {self.fakultet}.")
        elif self.spol == 'M' and bodovi < 45:
            print(f"Student {self.ime} {self.prezime} nije položio kolegij {kolegij} na fakultetu {self.fakultet}.")
```
