```
import turtle
turtle.colormode(1.0)
crvena=1
plava=zelena=0
for i in range(5):
    turtle.color(crvena, zelena, plava)
    turtle.begin_fill()
    for i in range(4):
        turtle.lt(90)
        turtle.fd(80)
    turtle.end_fill()
    crvena=crvena-0.1
    turtle.pu()
    turtle.fd(80)
    turtle.pd()
turtle.done()
```
