```
def ects(predmeti, polozeni):
    ukupno = 0
    for predmet in polozeni:
        if predmet in predmeti:
            ukupno += predmeti[predmet]
    print(f"Ukupan broj ECTS bodova: {ukupno}")
    return ukupno
```

```
ects(predmeti, polozeni)
```
