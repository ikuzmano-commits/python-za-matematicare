```
import pandas as pd
import matplotlib.pyplot as plt
```

```
# učitavanje CSV-a
url = ( 
"https://raw.githubusercontent.com/datasciencedojo/datasets/"
"master/titanic.csv"
)
data = pd.read_csv(url)
```

```
# pretvorba stupca Age u numerički tip
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

```
# filtriranje stupaca
procisceni_podaci = data[['Pclass','Sex','Age','SibSp','Parch','Fare','Survived']]
print(procisceni_podaci.head(6))
```

```
# uklanjanje putnika s dob < 1 i cijena karte < 10
procisceni_podaci = procisceni_podaci[(procisceni_podaci.Age >= 1) & (procisceni_podaci.Fare >= 10)]
print("Dimenzija filtriranog skupa:", procisceni_podaci.shape)
```

```
# filtriranje prve klase
prva_klasa = procisceni_podaci[procisceni_podaci.Pclass == 1]
```

```
# provjera različitih dobi
print(set(prva_klasa.Age))
```

```
# scatter plot: dob vs cijena karte, plavi kvadratići
plt.scatter(prva_klasa.Age, prva_klasa.Fare, color='blue', marker='s')
plt.xlabel('dob')
plt.ylabel('cijena karte')
plt.title('Cijene karata prema dobi - 1. klasa')
plt.show()
```

```
# otkrivanje outliera: najskuplja karta
max_fare = prva_klasa.Fare.max()
print(prva_klasa[prva_klasa.Fare == max_fare])
```

```
# uklanjanje outliera
prva_klasa = prva_klasa[prva_klasa.Fare != max_fare]
```

```
# ponovni scatter plot s vertikalno rotiranim x-osom
plt.scatter(prva_klasa.Age, prva_klasa.Fare, color='blue', marker='s')
plt.xlabel('dob')
plt.ylabel('cijena karte')
plt.title('Cijene karata prema dobi - 1. klasa')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('Cijene_karata_1_klasa.png')
plt.show()
```

```
# scatter plot cijena karte vs preživljavanje
plt.scatter(prva_klasa.Fare, prva_klasa.Survived)
plt.xlabel('cijena karte')
plt.ylabel('preživio (0/1)')
plt.title('Odnos cijene karte i preživljavanja')
plt.show()
```
