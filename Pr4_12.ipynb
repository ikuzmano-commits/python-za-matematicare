```
t = ('a', 'b', 'suzana', 'z', 4, 2, 'primjer')
```

```
# pristup elementima pomoću indeksa
print(t[0])     # prvi element n-torke
print(t[-1])    # posljednji element
print(t[1:3])   # elementi od indeksa 1 do 2
```

```
# provjera nalazi li se element u n-torki
if 'primjer' in t:
    print("'primjer' se nalazi u n-torki.")
```

```
# ručno pretraživanje pomoću petlje
for k in range(len(t)):
    if t[k] == 'primjer':
        print(f"Traženi element '{t[k]}' nalazi se na mjestu {k}.")
```
