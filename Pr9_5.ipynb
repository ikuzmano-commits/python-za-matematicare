```
import turtle
turtle.colormode(255)
turtle.begin_fill() # početak bojanja
turtle.fillcolor(0,0,255) # bojanje unutrašnjosti plavom bojom
turtle.circle(100)
turtle.end_fill() #kraj bojanja
turtle.pu()
turtle.goto(-190,-190)
turtle.pd()
turtle.pencolor(255,0,0) # postavi boju olovke u crvenu 
turtle.circle(100,120)
turtle.pu()
turtle.pencolor(0,0,0) # postavi boju olovke u crnu
turtle.goto(190,-110)
turtle.pd()
turtle.colormode(255)
turtle.begin_fill()
turtle.fillcolor(0,250,0) # oboj unutrašnjost u zelenu boju
turtle.circle(100,360,4)
turtle.end_fill()
turtle.done()
```
