```
# Unos brojnika i nazivnika
a = int(input("Unesite brojnik razlomka (a): "))
b = int(input("Unesite nazivnik razlomka (b): "))
```

```
# Provera da nazivnik nije nula
if b == 0:
    print("Nazivnik ne može biti nula.")
else:
    # Euklidov algoritam za pronalazak najvećeg zajedničkog delitelja (NZD)
    a0 = a  #U a0 i b0 spremamo originalne a i b
    b0 = b

    while b != 0:
        a, b = b, a % b
    
    nzd = a  # Nakon petlje, 'a' sadrži NZD

    # Skraćeni brojnik i nazivnik
    a1 = a0 // nzd
    b1 = b0 // nzd

    # Ispis rezultata
    print(f"Potpuno skraćeni razlomak {a0}/{b0} je {a1}/{b1}.")
```
