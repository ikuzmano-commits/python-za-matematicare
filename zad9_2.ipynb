```
from matplotlib import pyplot as plt
import random
```

```
br1=0
br2=0
br3=0
br4=0
br5=0
```

```
for i in range(0,50):
    broj=random.randint(0,100)
    if 0<=broj<=39:
        br1=br1+1
    elif 40<=broj<=59:
        br2=br2+1
    elif 60<=broj<=74:
        br3=br3+1
    elif 75<=broj<=89:
        br4=br4+1
    else:
         br5=br5+1
```

```
ocjene=[1,2,3,4,5]
bodovi=[br1,br2,br3,br4,br5]
```

```
plt.pie(bodovi,labels=ocjene)
```
