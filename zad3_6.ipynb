```
def tangenta(x0,a,b,c):
     y0=a*x0**2+b*x0+c   #račun ordinate točke 
     k=2*a*x0+b               #y-y0=k(x-x0), k=f'(x0)
     l=y0-k*x0
     print(f't... y={k}x+{l}')
     if k == 0:
        print("Normala je vertikalna.")
     else:
        print(f'n... y={-1/k}x+{y0+(1/k)*x0}')
```

```
tangenta(1,1,2,3)
```
