```
import matplotlib.pyplot as plt
import math
import numpy as np # uvodimo numpy kao np (kratko ime praktično za pisanje) 
x = np.arange(-10,10,0.01) # generiramo listu brojeva s početkom -10, krajem 10 i korakom 0.01
y= [math.sin(t) + t + t* math.sin(t) for t in x]
```

```
plt.plot(x,y,color='red')
```

```
plt.xlabel('x - os')
```

```
plt.ylabel('y - os')
```

```
plt.title('Graf funkcije sinx+x+xsinx')
# funkcija za prikaz slike
plt.show()
```
