```
class Trokut:
    def __init__(self, a,b=0,c=0):
        self.a=a
        if b==0:
            self.b=a
        else:
            self.b=b
        if c==0:
            self.c=self.b
        else:
            self.c=c
    def ispisi(self):
        print(f'({self.a}, {self.b}, {self.c})')
    def tip(self):
        if self.a == self.b == self.c:
            return 'jdnakostranican'
        elif self.a != self.b and  self.c != self.a and self.b != self.c:
            return 'raznostranican'
        else:
            return 'jednakokracan'
```

```
T=Trokut(3,2) # objekt T koji je jednakokračan trokut 
T.ispisi()
print(T.tip())
```

```
T1=Trokut(3,2,5)  # objekt T1 koji je raznostranični trokut
T1.ispisi()
print(T1.tip())
```
