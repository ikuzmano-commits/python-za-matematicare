```
def nzd(a, b):
    # ako je manji od brojeva 0, vraćamo veći broj
    if min(a,b) == 0:
        return max(a,b)
    else:
        # rekurzivno pozivanje funkcije prema Euklidovom algoritmu
        return nzd(b, a % b)
```

```
a = int(input("Unesite prvi broj (a): "))
b = int(input("Unesite drugi broj (b): "))
```

```
rezultat = nzd(a, b)
print(f"Najveći zajednički djelitelj brojeva {a} i {b} je {rezultat}.")
```
