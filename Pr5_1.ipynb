```
import numpy as np
```

```
# inicijalizacija matrica A i B pomoću np.array
A = np.array([[1, 2], [4, 5]])
B = np.array([[7, 8], [9, 10]])
```

```
# ispis matrice
print("A =", A) 
```

```
# i-ti redak matrice A dobija se naredbom A[i-1]
# indeksi redaka i stupaca počinju s 0 (slično kao za liste)
print("A[1] =", A[1])  # 2. redak od A 
print("A[0,1] =", A[0,1])   # element na poziciji (1,2)
```

```
# i-ti stupac od A se dobije naredbom A[:,i-1]
print("A[:,0] =",A[:,0]) # Prvi stupac
```

```
print(A==B) # ispitivanje jednakosti po elementima
np.array_equal(A,B) # vrati True ako matrice jednake, inače vrati False
np.allclose(A, B) #vrati True ako su odgovarajući elementi matrica dovoljno blizu jedan drugome, unutar zadane tolerancije.
```

```
# zbrajanje matrica 
C=A+B
print (C)
D=2*A+4*B
print(D)
```

```
# oduzimanje matrica
E=A-B
print(E)
```

```
# množenje matrica * i np.dot()
print ("Umnožak matrica A i B po elementima je : ")
print('A*B=',A*B) # svaki element matrice A množi s odgovarajućim elementom matrice B na istoj poziciji
print ("Umnožak matrica A i B  je : ")
print (np.dot(A,B)) # standardno matrično množenje dviju matrica
print(A @ B) # drugi način za standardno matrično množenje 
```

```
# suma svih elemenata matrice np.sum(A), po retcima np.sum(A,axis=1), po stupcima np.sum(A,axis=0)
print('suma elemenata matrice A je:')
print (np.sum(A))
print('sume  elemenata po retcima  matrice A su:')
print (np.sum(A,axis=1))
print('sume  elemenata po stupcima  matrice A su:')
print (np.sum(A,axis=0))
```

```
# transponiranje A.T
print (A.T) 
```
