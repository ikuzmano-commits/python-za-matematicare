```
import sympy as sp
```

```
x = sp.symbols('x')
```

```
f = -x**3 + 2*x**2 - x + 1
```

```
df = sp.diff(f, x) # prva derivacija funkcije f
print("f'(x) =", df)
```

```
x0 = 2
y0 = f.subs(x, x0).evalf()         # f(x0)
k= df.subs(x, x0).evalf() # nagib tangente
```

```
print(f"Vrijednost funkcije u x0={x0}: f(x0) = {y0}")
print(f"Nagib tangente u x0={x0}: k = {k}")
```

```
# jednadžba tangente: y = k*(x - x0) + f(x0)
tangent = k*(x - x0) + y0
print("Jednadžba tangente:", tangent)
```
