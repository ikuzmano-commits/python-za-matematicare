```
import sympy as sp
```

```
# definicija varijable i funkcije
x = sp.symbols('x')
f = sp.exp(x) - 2*x - 2
df = sp.diff(f, x)
```

```
# početna točka i broj iteracija
x0 = -1
n = 5
```

```
# Newtonova metoda
xk = x0
print(f"x0 = {xk}")
```

```
for k in range(1, n + 1):
    xk = xk - f.subs(x, xk) / df.subs(x, xk)
    print(f"x{k} = {sp.N(xk)}")
```
