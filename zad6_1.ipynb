```
import sympy as sp
```

```
# definicija simboličke varijable
x = sp.symbols('x')
```

```
# definicija funkcije koristeći simboličku eksponencijalnu funkciju
f = 2*sp.exp(-x) + x
```

```
print('f =', f)
```

```
# numerička evaluacija za zadane vrijednosti
for v in [0, 0.2, 0.3, 0.4]:
    a = f.subs(x, v)        # zamjena simboličke varijable numeričkom vrijednošću
    print(a.evalf())        # evalf daje decimalnu aproksimaciju
```
