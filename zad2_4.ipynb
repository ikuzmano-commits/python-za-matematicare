```
i1 = input('Igrač 1 - unesite k, s ili p: ')
i2 = input('Igrač 2 - unesite k, s ili p: ')
```

```
# provjera ispravnosti unosa
if i1 != 'k' and i1 != 's' and i1 != 'p':
    print(f'Igrač 1 nije napravio dobar unos, unio je "{i1}".')
elif i2 != 'k' and i2 != 's' and i2 != 'p':
    print(f'Igrač 2 nije napravio dobar unos, unio je "{i2}".')
else:
    # logika igre
    if i1 == i2:
        print('Nerješeno.')
    elif (i1 == 'k' and i2 == 's') or (i1 == 's' and i2 == 'p') or (i1 == 'p' and i2 == 'k'):
        print('Pobjedio je Igrač 1.')
    else:
        print('Pobjedio je Igrač 2.')
```
