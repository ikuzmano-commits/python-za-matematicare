```
import math
```

```
def f4(a, b):
    # 1. Provjera da nazivnik nije nula
    if b == 0:
        return("Nazivnik ne može biti nula.")

    # 2. Skraćivanje razlomka
    nzd = math.gcd(a, b)
    a //= nzd
    b //= nzd

    # 3. Uklanjanje faktora 2 i 5 iz nazivnika
    b0=b # u b0 sačuvamo originalni skraćeni razlomak
    while b % 2 == 0:
        b //= 2
    while b % 5 == 0:
        b //= 5

    # 4. Odlučivanje o vrsti decimalnog zapisa
    if b == 1:   #b je djeljiv samo s 2 i/ili 5
        print(f"Broj {a*nzd}/{b0*nzd} ima konačan decimalni zapis.")
    elif b==b0:  #nazivnik nije djeljiv ni s 2 ni s 5
        print(f"Broj {a*nzd}/{b0*nzd} ima beskonačan čisti periodički zapis.")
    else:     
        print(f"Broj {a*nzd}/{b0*nzd} ima beskonačan mješoviti decimalni zapis.")
```

```
f4(30,16)
f4(15,45)
f4(5,12)
```
