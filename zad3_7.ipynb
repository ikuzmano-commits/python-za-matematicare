```
import math
```

```
def skrati(brojnik, nazivnik):
    if nazivnik == 0:
        return "Nazivnik ne može biti nula!"
    
    # Izračun najvećeg zajedničkog djelitelja (NZD) koristeći funkciju gcd iz math biblioteke
    nzd = math.gcd(brojnik, nazivnik)
    
    brojnik1 = brojnik // nzd
    nazivnik1 = nazivnik // nzd
    
    # Ako je nazivnik1 negativan, prebacujemo negativni predznak u brojnik1
    if nazivnik1 < 0:
        brojnik1 = -brojnik1
        nazivnik1 = -nazivnik1
    
    return brojnik1,nazivnik1
```

```
(b1,n1) = skrati(34, -12)
print(f"Skraćeni razlomak je: {b1}/{n1}")
```
