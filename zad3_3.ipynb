```python
def potencija3(broj):
    if (broj <= 0):
        return False
    while (broj % 3 == 0):
        broj //= 3 #broj=broj//3
    return broj == 1
```

```python
print(potencija3(5))
print(potencija3(9))
```

```python
#drugi način
def potencija3b(broj):
    import math
    if (broj <= 0):
        return False
    pot=math.log(broj,3)
    return (int(pot)==pot)
```

```python
print(potencija3b(5))
print(potencija3b(9))
```
