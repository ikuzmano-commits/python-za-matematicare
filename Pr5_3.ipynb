```
import numpy as np
```

```
# jediniƒçna matrica
A=np.eye(3)
print('A=',A)
```

```
# nul matrica 
B=np.zeros((2,3))
print('B=',B)
```

```
# matrica jedinica
C=np.ones((3,3))
print('C=',C)
```

```
# dijagonalna matrica
D= np.diag([3, -4, 13, 9])
print('D=',D)
```

```
# random matrica
R=np.random.rand(3,2) # elementi iz intervala (0,1)
print('R=',R)
```

```
R2=np.random.randint(5, size=(2, 4))
print('R2=',R2)
R3=np.random.randint(13,25, size=(2, 2))
print('R3=',R3)
```
