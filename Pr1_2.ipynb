```python
#Unos vrijednosti varijable s tipkovnice: ime_varijable=tip_varijable(input('Poruka'))

#unos vrijednosti int varijable s tipkovnice
a=int(input('Unesi neki cijeli broj:'))

#unos vrijednosti float varijable s tipkovnice
b=float(input('Unesi neki realni broj:'))

#unos vrijednosti string varijable s tipkovnice
#!(string je default tip za naredbu input)
c=input('Unesi naziv svog omiljenog kolegija:')

print('Varijabla a ima vrijednost',a,'.')
print('Kolegij',c,'jako je zanimljiv.')
print('a=',a,'i b=',b)


# f-stringovi mogu biti lakši i čitljiviji način umetanja varijable u string

print(f"Varijabla a ima vrijednost {a}.")
print(f"Kolegij {c} jako je zanimljiv.")
print(f"a={a} i b={b}")
```
