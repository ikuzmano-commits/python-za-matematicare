```
import numpy as np
```

```
def isInverz(A, B):
    n = A.shape[0]
    I = np.eye(n)
    return np.allclose(A @ B, I) #provjera A*B==I do na numeriƒçku toleranciju
```

```
A = np.random.random((3,3))
B = np.random.random((3,3))
print(isInverz(A,B))
```

```
A2 = np.array([[1, 2],[-2, 1]])
B2 = np.array([[0.2, -0.4],[0.4, 0.2]])
print(isInverz(A2, B2))
```
