```
def suma(a, b):     # funkcija koja vraća zbroj a+b
    return a + b
```

```
c = suma(2, 4)       # c poprima vrijednost 6 (tip int) i može se koristiti za daljnji račun
print(c)             # ispisuje 6
```

```
def fun(a, b, c):    # funkcija koja nema return
    print(a*b + b*c + c*a)
```

```
fun(1, 2, 3)         # ispisuje 11, ali se fun(1,2,3) ne može koristiti u daljnjem računu
```

```
d=fun(1,2,3)
print(d)           # d ima vrijednost None
print(type(d))     # 'NoneType'
```
