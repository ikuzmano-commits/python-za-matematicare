```
# definiranje rječnika
ucenici = {"Ana": 5, "Ivan": 4, "Mara": 5, "Pero": 3}
```

```
# ispis cijelog rječnika
print(ucenici)
```

```
# ispis vrijednosti za određenog učenika
print("Ocjena učenice Ane je:", ucenici["Ana"])
```

```
# dodavanje novog elementa
ucenici["Luka"] = 4
print(ucenici)
```

```
# promjena vrijednosti postojećeg ključa
ucenici["Pero"] = 4
print(ucenici)
```
