```
lista = ['prvi', 'a', 4, -4.6, True]   # elementi liste mogu biti različitih tipova podataka
print(lista)
```

```
print(lista[2])    # ispis elementa s indeksom 2
lista[2] = 7       # promjena vrijednosti elementa na indeksu 2
print(lista)
```

```
print(lista[-2])   # negativni indeks označava brojanje od kraja (drugi element s kraja)
```
