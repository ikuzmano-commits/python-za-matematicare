```
def struktura(A):
    m, n = A.shape
    if m != n:
        print("Matrica nije kvadratna.")
        return
    diag=True
    for i in range(0,n):
        for j in range(0,n):
            if i!=j and A[i][j]!=0:
                diag=False
    if diag==True:
        print('Matrica je dijagonalna!')
    gornjetr=True
    for i in range(0,n):
        for j in range(0,i):
            if  A[i][j]!=0:
                gornjetr=False
    if gornjetr==True:
        print('Matrica je gornjetrokutasta!')
    donjetr=True
    for i in range(0,n):
        for j in range(i+1,n):
            if  A[i][j]!=0:
                donjetr=False
    if donjetr==True:
        print('Matrica je donjetrokutasta!')
    tridiag=True
    for i in range(0,n):
        for j in range(0,i-1):
            if  A[i][j]!=0:
                tridiag=False
        for j in range(i+2,n):
            if  A[i][j]!=0:
                tridiag=False
    if tridiag==True:
        print('Matrica je tridijagonalna!')
    if diag==False and gornjetr==False and donjetr==False and tridiag==False:
        print('Matrica nema strukturu!')
```
