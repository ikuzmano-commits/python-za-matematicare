```
import turtle
turtle.circle(100) # crtanje kružnice zadanog radijusa
turtle.pu() # podizanje olovke
turtle.goto(-190,-190) # premještanje olovke
turtle.pd() # spuštanje olovke za pisanje
turtle.circle(100,120) # crtanje kružnog isječka tako da mu se zada radijus i kut
turtle.pu()
turtle.goto(190,-110) # premještanje olovke
turtle.pd()
turtle.circle(100,360,4) # crtanje pravilnog n-terokuta, zadaje se radijus , kut i broj vrhova
turtle.done()
```
