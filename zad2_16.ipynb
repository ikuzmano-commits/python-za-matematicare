```
# unos znamenke k od korisnika
k = int(input("Unesite znamenku k: "))
```

```
# provjera je li unesena znamenka u rasponu od 0 do 9
if k < 0 or k > 9:
    print("Molimo unesite znamenku u rasponu od 0 do 9.")
else:
    print(f"Troznamenkasti brojevi kojima je zadnja znamenka {k} su:")

    # petlja za prolazak kroz sve troznamenkaste brojeve
    for stot in range(1, 10):
        for dest in range(0, 10):
            broj = stot * 100 + dest * 10 + k
            print(broj)
```
