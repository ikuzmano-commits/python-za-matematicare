```python
#Pridruživanje vrijednosti varijabli. Tip varijable ne mora se eksplicitno navoditi.
a=5              #varijabla tipa int
print(a)         #ispis vrijednosti
print(type(a))   #ispis tipa varijable

b=3.5            #varijabla tipa float
print(b)
print(type(b))

c='kaktus'       #varijabla tipa string, koriste se jednostruki ili dvostruki navodnici
print(c)
print(type(c))

d=True           #varijabla tipa bool
print(d)
print(type(d))
```
