```
import math
```

```
def maxudaljenost(tocke):
    max_d = 0                   # početno postavljamo maksimalnu udaljenost
    par = (None,None)      # i par točaka koji ćemo kasnije popuniti

    # prolaz kroz sve parove točaka
    for i in range(len(tocke)):
        for j in range(i + 1, len(tocke)):
            x1, y1 = tocke[i]
            x2, y2 = tocke[j]
            d = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)  # Euklidska udaljenost

            if d > max_d:        # ako je pronađena veća udaljenost, ažuriramo maksimum
                max_d = d
                par = (tocke[i], tocke[j])

    return (par[0], par[1], max_d)
```

```
# primjer poziva
tocke = [(2, 3), (1, 5), (4, -2), (0, 0), (3, 3)]
rezultat = maxudaljenost(tocke)
```

```
print(f"Najudaljenije točke su {rezultat[0]} i {rezultat[1]}")
print(f"Njihova udaljenost iznosi {rezultat[2]:.3f}")
```
