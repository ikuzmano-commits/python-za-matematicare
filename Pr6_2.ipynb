```
import sympy as sp
```

```
x = sp.symbols('x')
```

```
f = x**3 - 5*x**2 + 2*x - 7
```

```
df = sp.diff(f, x)
d2f = sp.diff(f, x, 2) #druga derivacija
```

```
print("f =", f)
print("f' =", df)
print("f'' =", d2f)
```

```
x, y = sp.symbols('x y')
```

```
f = x**2*y + sp.sin(x*y)
```

```
fx = sp.diff(f, x) #parcijalne derivacije
fy = sp.diff(f, y) #parcijalne derivacije
```

```
print("f_x =", fx)
print("f_y =", fy)
```
