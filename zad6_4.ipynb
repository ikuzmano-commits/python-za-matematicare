```
import sympy as sp
```

```
x = sp.symbols('x')
f = 2*x**3 + 3*x**2 - 12*x + 1
```

```
# prva i druga derivacija
df = sp.diff(f, x)
ddf = sp.diff(df, x)
```

```
# kritične točke
nul = sp.solve(df, x)
nul = sorted([float(c) for c in nul])
print("Kritične točke:", nul)
```

```
# intervali monotonosti
intervali = [(-sp.oo, nul[0])] + \
            [(nul[i], nul[i+1]) for i in range(len(nul)-1)] + \
            [(nul[-1], sp.oo)]
```

```
print("\nIntervali monotonosti:")
for a, b in intervali:
    if a == -sp.oo:
        test = b - 1
    elif b == sp.oo:
        test = a + 1
    else:
        test = (a + b) / 2

    sign = sp.sign(df.subs(x, test))
    monotonost = "rastući" if sign > 0 else "padajući"
    print(f"Interval ({a}, {b}): f(x) {monotonost}")
```

```
# lokalni ekstremi 
print("\nLokalni ekstremi:")
for cp in nul:
    f_val = f.subs(x, cp)
    f2_val = ddf.subs(x, cp)
    if f2_val > 0:
        tip = "minimum"
    elif f2_val < 0:
        tip = "maksimum"
    else:
        tip = "test neodređen"
    print(f"x = {cp}: f(x) = {f_val} ({tip})")
```
