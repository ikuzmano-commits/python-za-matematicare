```
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```
x = sp.symbols('x') # simboli훾ka varijabla
```

```
f = x**3 - 3*x**2 + 2
```

```
df = sp.diff(f, x) # prva derivacija
print("f'(x) =", df)
```

```
x0 = 0.5
k = df.subs(x, x0).evalf() # vrijednost derivacije u to훾ki x0
y0 = f.subs(x, x0).evalf() #vrijednost funkcije u to훾ki x0
print(f"Nagib tangente u x={x0}: {k}")
print(f"Vrijednost funkcije u x={x0}: {y0}")
```

```
tang = k*(sp.Symbol('x') - x0) + y0 #tangenta
```

```
# priprema podataka za crtanje s NumPy
x_vals = np.linspace(-1, 3, 400)  # interval
f_vals = np.array([f.subs(x, xi).evalf() for xi in x_vals])
tang_vals = np.array([tang.subs(sp.Symbol('x'), xi).evalf() for xi in x_vals])
```

```
# crtanje funkcije i tangente
plt.figure(figsize=(8,5))
plt.plot(x_vals, f_vals, label="f(x) = x^3 - 3x^2 + 2", color='blue')
plt.plot(x_vals, tang_vals, label=f"Tangent u x={x0}", color='red', linestyle='--')
plt.scatter([x0], [y0], color='black')  # to훾ka tangente
plt.title("Funkcija i tangenta u x0")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
