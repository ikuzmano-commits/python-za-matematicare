```
import math
```

```
def sredine(lista):
    suma = 0
    produkt = 1
    
    for x in lista:
        suma += x
        produkt *= x
    
    n = len(lista)
    aritmeticka = suma / n
    geometrijska = math.pow(produkt, 1/n)
    
    return aritmeticka, geometrijska   # funkcija vraća obje sredine 
```

```
brojevi = [5, 6, 8, 1, 9, 11, 12, 3]
asr, gsr = sredine(brojevi)
print(f"Aritmetička sredina je: {asr:.2f}")
print(f"Geometrijska sredina je: {gsr:.2f}")
```
