```
def brojac(rijec):
    rj = {}  # prazan rječnik u kojem će se spremati parovi (slovo : broj_pojavljivanja)
    
    for slovo in rijec:
        # rj.get(slovo, 0) vraća trenutačni broj pojavljivanja slova,
        # a ako slovo još nije u rječniku, vraća 0
        rj[slovo] = rj.get(slovo, 0) + 1

    # pretvaramo rječnik u listu uređenih parova (ključ, vrijednost)
    lista = list(rj.items())
    
    # sortiramo listu po abecednom redoslijedu ključeva
    lista.sort()
    
    return lista
```

```
# primjer poziva
print(brojac("matematika"))
```
