```
import random
```

```
def slpol(lista):
    n = len(lista)
    k = n // 2                        # broj elemenata koje biramo
    nova = random.sample(lista, k)  # nasumično biranje bez ponavljanja
    print("Slučajno odabrani elementi:", nova)
```

```
brojevi = [5, 6, 1, 9, 11, 12, 3, 4]
slpol(brojevi)
```
