```
import numpy as np
```

```
def dd(A):
    m, n = A.shape

    if m != n:
        print("Matrica nije kvadratna.")
        return False

    for i in range(n):
        suma = 0
        for j in range(n):
            if i != j:
                suma += abs(A[i, j])

        if abs(A[i, i]) <= suma:
            return False

    return True
```

```
A=np.array([[-10, 1, 7],[5, 10, 2],[-5,-2, 10]])
print(dd(A))
```

```
B=np.random.randint(-1,1, size=(3,3))
print(dd(B))
```
