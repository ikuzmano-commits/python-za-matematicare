```
import numpy as np
```

```
def isSim(A):
    B = A.T
    return np.array_equal(A,B) #jo≈° bolje koristiti np.allclose(A,B) 
```

```
def isAsim(A):
    B = -A.T
    return  np.array_equal(A,B)
```

```
A = np.array([[1, 2],[2, 3]])
B = np.array([[0.2, -0.4],[0.4, 0.2]])
print(isSim(A))
print(isAsim(B))
```
