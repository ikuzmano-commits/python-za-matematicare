```
# funkcija za pronalazak koeficijenata pravca kroz točke A i B
def prav2t(x1, y1, x2, y2):
    if x1 == x2:
        return None
    k = (y2 - y1) / (x2 - x1)
    l = y1 - k * x1
    return k, l
```

```
# funkcija koja ispisuje jednadžbu sekante kroz točke na paraboli f(x) = ax^2 + bx + c
def sekanta_kvadratna(a, b, c, x1, x2):
    # izračunavanje ordinata za točke na paraboli
    y1 = a * x1**2 + b * x1 + c
    y2 = a * x2**2 + b * x2 + c
    
    # dobivanje koeficijenata za sekantu koristeći prav2t funkciju
    rez=prav2t(x1, y1, x2, y2)
    if rez is None:
        print("Pravac je vertikalan.")
    else:
        k, l = rez
        print(f"Jednadžba sekante kroz točke s apscisama {x1} i {x2} je: y = {k:.2f}x + {l:.2f}")
```

```
a, b, c = 1, -3, 2  # koeficijenti kvadratne funkcije f(x) = x^2 - 3x + 2
x1, x2 = -1, 2       # apscise točaka kroz koje sekanta prolazi
```

```
sekanta_kvadratna(a, b, c, x1, x2)
```
