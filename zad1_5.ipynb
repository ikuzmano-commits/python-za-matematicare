```python

a = float(input('Unesi pozitivan realan broj a: '))
b = float(input('Unesi pozitivan realan broj b: '))
c = float(input('Unesi pozitivan realan broj c: '))
```

```python
# izračun aritmetičke i geometrijske sredine
aritmeticka_sredina = (a + b + c) / 3
geometrijska_sredina = (a * b * c) ** (1 / 3)
```

```python
# ispis rezultata
print(f'Aritmetička sredina je: {aritmeticka_sredina:.2f}')
print(f'Geometrijska sredina je: {geometrijska_sredina:.2f}')
```
