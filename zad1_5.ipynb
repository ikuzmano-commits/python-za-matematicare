```
a = float(input('Unesi a: '))
b = float(input('Unesi b: '))
c = float(input('Unesi c: '))
```

```
# izračun aritmetičke i geometrijske sredine
aritmeticka_sredina = (a + b + c) / 3
geometrijska_sredina = (a * b * c) ** (1 / 3)
```

```
# ispis rezultata
print(f'Aritmetička sredina je: {aritmeticka_sredina}')
print(f'Geometrijska sredina je: {geometrijska_sredina}')
```
