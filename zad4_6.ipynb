```
# S1 = {x + 2 : x iz {-1, 0, 1, 2, 3}}
l1 = [-1, 0, 1, 2, 3]
s1 = [x + 2 for x in l1]
print("S1 =", s1)
```

```
# S2 = {x iz {1, 2, ..., 50} takav da je x neparan}
s2 = [x for x in range(1, 51) if x % 2 != 0]
print("S2 =", s2)
```

```
# S3 = {x1^2 : (x1, x2) iz {(0,0), (1,0), (15,10), (12,4)}}
l3 = [(0, 0), (1, 0), (15, 10), (12, 4)]
s3 = [x1**2 for (x1, x2) in l3]
print("S3 =", s3)
```

```
# S4 = A Ã— B, A = {1, 2, 3}, B = {2, 3, 4}
A = [1, 2, 3]
B = [2, 3, 4]
s4 = [(x, y) for x in A for y in B]
print("S4 =", s4)
```
