```
lista = ['prvi', 'a', 4, -4.6, True]
```

```
# dodavanje elemenata
lista.append(10)            # dodaje element 10 na kraj liste
print(lista)
```

```
lista.append([11, 12])      # dodaje novu listu kao jedan element na kraj postojeće
print(lista)
```

```
# uklanjanje  elemenata
a=lista.pop() 
print(a)   
print(lista)             # uklanja zadnji element liste i sprema ga u varijablu a
lista.pop()                 # uklanja zadnji element liste bez  da ga sačuva
print(lista)
```

```
# brisanje određenog elementa po vrijednosti
lista.remove(-4.6)          # briše element -4.6 iz liste
print(lista)
```

```
# ako se element pojavljuje više puta, briše se samo njegovo prvo pojavljivanje
lista1 = ['prvi', True, 'a', 4, -4.6, True]
lista1.remove(True)
print(lista1)
```
