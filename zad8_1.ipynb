```
class Tocka:
    # konstruktor koji će početne vrijednosti postaviti na 0 ukoliko se ne unesu
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y= y
    # metoda koja vraća udaljenost točke do ishodišta
    def udaljenost(self):
        return (self.x**2+self.y**2)**0.5
    # metoda koja vraća udaljenost ove točke do neke druge točke T
    def udaljenostDoTocke(self, T):
        return ((self.x-T.x)**2+(self.y-T.y)**2)**0.5
    def ispisi(self):
        print(f"({self.x},{self.y})")
```

```
T2 = Tocka(3,4) # objekt T2 koji je točka (3,4)
print(T2.udaljenost()) # udaljenost točke T2 do ishodišta
```

```
T3 = Tocka(y=6) # objekt T3 koji je točka (0,6)
T3.ispisi()
print(T3.udaljenost()) # udaljenost točke T3 do ishodišta
print(T3.udaljenostDoTocke(T2)) # udaljenost između točaka T3 i T2
```

```
T1 = Tocka(9) # objekt T1 koji je točka (9,0) 
print(T1.udaljenost()) # udaljenost T1 do ishodišta
```
