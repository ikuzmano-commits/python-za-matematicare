```
def is_samoglasnik(slovo):
    samoglasnici = ['a','e','i','o','u','A','E','I','O','U']
    if slovo in samoglasnici:
        return True
    else:
        return False
```

```
rec = open('pjesma.txt','r')
bezs = open('bezsamoglasnika.txt','w')
```

```
for line in rec:
    lista=[]
    for element in line:
        if is_samoglasnik(element)==False:
            lista.append(element)

    # Isto ali  s list comprehension
   # lista = [element for element in line if not is_samoglasnik(element)]

   #bezs.write(str(lista)) # los ispis, trebalo bi zdruziti znakove
    bezs.write(''.join(lista))         #list1 = ['a', 'b', 'c', 'd', 'e']
                                         #print(''.join(list1)) vraca abcde
rec.close()
bezs.close()
```
