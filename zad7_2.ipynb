```
# Zadatak pod a)
import random
lista=range(5,21)
slucajno=random.choices(lista,k=5)
sl=open('slodabrani.txt','w')
sl.write(str(slucajno))
```

```
max=slucajno[0]
zbroj=0
for broj in slucajno:
    zbroj=zbroj+broj
    if broj>max:
        max=broj
```

```
sl.write('\n Najveci od svih slucajno odabranih brojeva je:')
sl.write(str(max))
```

```
sl.write('\n Zbroj svih slucajno odabranih brojeva je:'+str(zbroj) )
print(slucajno)
sl.close()
```

```
# Zadatak pod b)
import random
lista=range(2,101)
sl1=open('slbr.txt','w')
for i in range(1,11):
    slucajno=random.choices(lista,k=10)
    sl1.write(str(slucajno))
    zbroj=0
    for broj in slucajno:
        zbroj=zbroj+broj
    arsred=zbroj/10
    sl1.write('\n Aritmeticka sredina je:'+str(arsred)+'\n')
sl1.close()
```
