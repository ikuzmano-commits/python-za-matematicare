```
import math
broj=5
suma=0
br=0
while broj!=0 and math.log2(broj).is_integer()!=True: #ovo znači da je log2(n) nije cijeli broj, tj. da j n nije potencija broja 2. Može i kao  int(math.log2(n))!=math.log2(n)
    broj=int(input('Unesite broj:'))
    suma+=broj
    br+=1
arsr=suma/br
print(f'Aritmetička sredina unešenih brojeva je {arsr}.')
```
