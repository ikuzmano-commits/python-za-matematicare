```
import pandas as pd
data={'razred':[1, 2, 3, 4],
      'brucenika':[46, 30, 26, 46],
      'maksbod': [70, 100, 100, 95]}
# napravim padnas Data frame od danih podataka
df=pd.DataFrame(data)
print(df)
# dimenzija DataFrame
print(df.shape)
```

```
# manipuliranje recima i stupcima -> Series objekt
print(df.loc[1])
print(df['razred'])
```

```
# postavljanje uvjeta i ispitivanje uvjeta
uvjet=df['razred']<3
print(uvjet)
```

```
# filtriranje obzirom na uvjet
bod=df['maksbod']==100
print(bod)
```

```
df1=df[bod] # formiranje novog DataFrame od df koji će sadržavati one retke koji zadovoljavaju zadano svojstvo
print(df1)      
```
