```
class Plovilo:

    def __init__(self, naziv, duljina, kilometraza):
        self.naziv = naziv
        self.duljina = duljina
        self.kilometraza = kilometraza

    def plovi(self, milje):
        self.kilometraza += milje
        print(f"Nakon pređenih {milje} nautičkih milja, stanje brojača je: {self.kilometraza} milja.")
```

```
class Brod(Plovilo):
   
    def __init__(self, naziv, duljina, kilometraza, nosivost):
        super().__init__(naziv, duljina, kilometraza)
        self.nosivost = nosivost

    def utovari(self, tona):
        print(f"Na brod je utovareno {tona} tona tereta.")
```

```
class Jahta(Plovilo):
    def __init__(self, naziv, duljina, kilometraza, vlasnik):
        super().__init__(naziv, duljina, kilometraza)
        self.vlasnik = vlasnik

    def party(self):
        print(f"Na jahti vlasnika {self.vlasnik} se održava zabava!")
```

====== Primjeri korištenja ======

```
# Primjer 1: Brod
b1 = Brod("Adria", 50, 1200, 2000)
b1.utovari(350)
b1.plovi(22)
```

```
# Primjer 2: Jahta
j1 = Jahta("Bella", 35, 800, "Ivan Horvat")
j1.party()
j1.plovi(8)
```
