```
# Definiranje klase Arhiva
class Arhiva:
    def __init__(self):
        self.podaci=[] # prazna lista
    # metoda za dodavanje elementa u listu
    def dodaj(self,x):
        self.podaci.append(x)
    # metoda za dodavanje istog elementa dva puta
    def dodaj2puta(self,x):
        self.dodaj(x) # poziva metodu dodaj prvi put
        self.dodaj(x) # poziva metoru dodaj drugi put
```

```
Arh=Arhiva() # stvaranje objekta Arh
Arh.dodaj('nesto') # u objekt se dodaje element 'nesto'
Arh.dodaj2puta('nesto2') # u objekt se dvaput dodaje element 'nesto2'
print(Arh.podaci)
```
