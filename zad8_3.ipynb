```
import math
class Kompleksan:
    def __init__(self, x,y):
        self.x=x
        self.y=y
    def ispisi(self):
        if self.y>0:
            print(f'z={self.x} + {self.y} i')
        else:
            print(f'z={self.x} - {abs(self.y)} i')     
    def zbroj(self, z1):
        re=(self.x + z1.x)
        im=(self.y + z1.y)
        zbr=Kompleksan(re,im)
        return zbr.ispisi()
    def umnozak(self, z1):
        re = self.x * z1.x - self.y * z1.y
        im = self.x * z1.y + self.y * z1.x
        um=Kompleksan(re,im)
        return um.ispisi()
    def kvocijent(self, z1):
        nazivnik = z1.x**2 + z1.y**2
        if nazivnik == 0:
            print('Dijenjelje s nulom nije dozvoljeno')
            return None
        re = (self.x * z1.x + self.y * z1.y) / nazivnik
        im = (self.y * z1.x - self.x * z1.y) / nazivnik
        return Kompleksan(re, im).ispisi()
    def modul(self):
        return (self.x**2+self.y**2)**0.5
    def argument(self):
        arg= math.atan2(self.y,self.x) #atan2 vraća kut iz [-pi,pi>
        if arg < 0:
            arg = 2 * math.pi + arg  # pretvori u [0, 2π)
        return arg
```

```
z=Kompleksan(-1,-1)
z.ispisi()
z1=Kompleksan(2,3)
z.zbroj(z1)
z.kvocijent(z1)
z.umnozak(z1)
print(z.modul())
print(z.argument())
```
