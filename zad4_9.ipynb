```
def najdulji(lista):
    najd = ""       # početno prazan string
    duljina = 0         # početna duljina
    
    for s in lista:
        if len(s) > duljina:
            duljina = len(s)
            najd = s
    
    print(f"Najdulji string je '{najd}' s duljinom {duljina}.")
    return najd, duljina   # funkcija vraća obje vrijednosti
```

```
dani = ['ponedjeljak', 'utorak', 'srijeda', 'cetvrtak', 'petak', 'subota', 'nedjelja']
```

```
najdulji(dani)
```
