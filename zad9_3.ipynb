```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
# funkcija najvećeg cijelog broja (floor)
def nc(x):
    return math.floor(x)
```

```
# decimalni dio
def dd(x):
    return x - nc(x)
```

```
# zadana funkcija
def f(x):
    return x**2 - nc(x)**2
```

```
# interval
x = np.arange(-10, 10, 0.001)
```

```
# vrijednosti funkcija
y = [f(t) for t in x]
y1 = [dd(t) for t in x]
```

```
# crtanje
plt.plot(x, y, label='f(x) = x² − nc(x)²')
plt.plot(x, y1, label='dd(x)')

plt.xlabel('x')
plt.ylabel('vrijednost')
plt.title('Graf funkcija f i dd')
plt.grid(True)
plt.legend()

plt.show()
```
