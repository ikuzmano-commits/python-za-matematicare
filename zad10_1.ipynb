```
# učitavanje potrebnih biblioteka
import pandas as pd
import matplotlib.pyplot as plt
```

```
# učitavanje podataka
url = (
    "https://raw.githubusercontent.com/resbaz/"
    "r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
)
podaci = pd.read_csv(url)
```

```
# pregled podataka
print(podaci.shape) # dimenzija tablice
print(podaci.head(10)) # prvih 10 redaka tablice
```

```
# različiti kontinenti
kontinenti = set(podaci['continent'])
print(kontinenti)
```

```
# čišćenje podataka (uklanjanje NaN vrijednosti)
podaci = podaci.dropna()
print(podaci.shape)
```

```
# filtriranje europskih država
europa = podaci[podaci['continent'] == 'Europe']
```

```
# godine u Europi
godine = sorted(list(set(europa['year'])))
print(godine)
```

```
# broj država po godinama
brDrzava = [europa[europa['year'] == g].shape[0] for g in godine]
akumuliraniBroj = []
suma = 0
for broj in brDrzava:
    suma += broj
    akumuliraniBroj.append(suma)
```

```
# Graf akumuliranog broja država
plt.figure(figsize=(10,5))
plt.plot(godine, akumuliraniBroj, marker='o')
plt.title("Broj europskih država kroz godine")
plt.xlabel("godine")
plt.ylabel("broj država")
plt.xlim(1950, 2010)
plt.grid(True)
plt.show()
```

```
#  histogram BDP-a po stanovniku (gdpPercap) za europske države
plt.figure(figsize=(10,5))
plt.hist(europa['gdpPercap'], bins=5, range=(0,65000), edgecolor='black')
plt.title("Distribucija BDP-a po stanovniku (Europa)")
plt.xlabel("GDP per capita ($)")
plt.ylabel("Broj država")
plt.show()
```

```
# kumulativni histogram broja država kroz godine
plt.figure(figsize=(10,5))
plt.hist(akumuliraniBroj, bins=len(godine), cumulative=True, edgecolor='black')
plt.title("Broj europskih država kroz godine")
plt.xlabel("godine")
plt.ylabel("kumulativni broj država")
plt.show()
```
