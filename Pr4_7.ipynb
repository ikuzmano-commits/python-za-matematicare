```
def obrni(l):
    l2 = []                     # nova, prazna lista
    while len(l) != 0:          # dok lista nije prazna
        l2.append(l.pop())      # uklanja zadnji element iz l i dodaje ga u l2
    return l2                   # funkcija vraÄ‡a obrnuti redoslijed elemenata
```

```
# poziv funkcije s listom kao argumentom
lista=[1, 2, 3, 4, 5]
print(obrni(lista))
print(lista) # kako funkcija mijenja originalnu listu, ona je na kraju prazna
```
