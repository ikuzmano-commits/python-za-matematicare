```
# otvaranje datoteke za čitanje
dat1 = open('pr1.txt', 'r')
print(dat1.read())        # čita se cijeli sadržaj datoteke
dat1.close()              # obavezno zatvaranje nakon čitanja
```

```
# primjer čitanja redak po redak
dat1 = open('pr1.txt', 'r')
redak1 = dat1.readline()  # čita se prvi redak
print(redak1)
redak2 = dat1.readline()  # čita se drugi redak
print(redak2)
dat1.close()
```

```
# primjer dodavanja sadržaja u datoteku
dat1 = open('pr1.txt', 'a')     # 'a' znači append (dodavanje)
dat1.write('Treci redak\n')     # upis novog retka u datoteku
dat1.close()
```

```
# čitanje svih redaka u obliku liste
dat1 = open('pr1.txt', 'r')
print(dat1.readlines())         # svaki redak je jedan element liste
dat1.close()
```
